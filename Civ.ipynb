```python
# !/usr/bin/python3
# -*- coding: utf-8 -*-
 
#Milosza
#make jupiter cell wider
import os
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

from tkinter import (Tk, RIDGE, Frame, Button, Label, IntVar, StringVar, Entry, Text, PhotoImage, END)
import numpy as np
import random as rand
from tkinter import scrolledtext
from PIL import Image, ImageTk

#constans
PATH = os.getcwd()
background = 'SkyBlue2'
buttons = "DeepSkyBlue2"
attack = os.path.join(PATH, 'attack.png')

kultura_color = 'dark orchid'
nauka_color = 'dodger blue'
sila_color = 'indian red'
cywilne_color = 'white'
militarne_color = 'orange red'
blue_dice_color = 'blue'
kamienie_color = 'gray55'
yellow_dice_color = 'orange'
zboze_color = 'gold'
robotnicy_color = 'orange'
zadowolenie_color = 'dark orange'

#Frame
root = Tk();        
root.title('Civilization VI 2.0')
Frame(root, width=1250, height=600, bg=background, relief="groove").pack()

class frame():
    def __init__(self, path, set_x, set_y, size_x, size_y):
        image = Image.open(path)
        image = image.resize((size_x, size_y), Image.ANTIALIAS) #The (50, 50) is (height, width)
        photo = ImageTk.PhotoImage(image)
        label = Label(image=photo, bg=background)
        label.image = photo # keep a reference!
        label.place(x=set_x, y=set_y)

def kwargs(arg):
    return {
    'set_x': arg,
    'set_y': 15,
    'size_x': 210,
    'size_y': 570}

frame(os.path.join(PATH, 'red_frame.png'), **kwargs(165))
frame(os.path.join(PATH, 'blue_frame.png'), **kwargs(375))
frame(os.path.join(PATH, 'green_frame.png'), **kwargs(585))
frame(os.path.join(PATH, 'orange_frame.png'), **kwargs(795))
        
culture = frame(os.path.join(PATH, 'culture.png'), 120, 35, 30, 30)
science = frame(os.path.join(PATH, 'science.png'), 120, 75, 30, 30)
attack = frame(os.path.join(PATH, 'atack.png'), 120, 120, 25, 25)
civil = frame(os.path.join(PATH, 'civil.png'), 120, 200, 30, 30)
tank = frame(os.path.join(PATH, 'tank.png'), 120, 230, 40, 40)
#bank
dice = frame(os.path.join(PATH, 'dice.png'), 120, 310, 40, 40)
stone = frame(os.path.join(PATH, 'stone.png'), 120, 360, 30, 30)
wheat = frame(os.path.join(PATH, 'wheat.png'), 120, 430, 45, 45)
dice2 = frame(os.path.join(PATH, 'dice2.png'), 120, 395, 40, 40)
peasant = frame(os.path.join(PATH, 'peasant.png'), 120, 475, 40, 40)
#happyness
happy = frame(os.path.join(PATH, 'happy.png'), 125, 520, 30, 30)
homer = frame(os.path.join(PATH, 'homer.png'), 1050, 400, 170, 170)

player_turn = StringVar()
player_turn.set('czerwony')
player_move = IntVar()
player_move.set(0)
button_state = StringVar()
button_state.set('disabled')

class util():        
    
    def update_button(self, button):
        
        users = {'czerwony':     'red',
                 'niebieski':    'blue',
                 'zielony':      'green',
                 'pomaranczowy': 'orange'}
        
        button.configure(bg = users.get(player_turn.get()))    
    
    def add(self, attribute, mod):                       #x = 1 if mod == 'add' else -1
        if mod == 'add':
            x = 1
        elif mod == 'subtrack':
            x = -1
        else:
            print('ERROR')

        attribute.set(attribute.get() + x)
        
    def turn(self):
        
        play = {'czerwony': czerwony,
               'niebieski': niebieski,
                 'zielony': zielony,
            'pomaranczowy': pomaranczowy}
        
        player = play.get(player_turn.get())
        return player        
    
    def end(self):
        
        play = {'czerwony': 'niebieski',
               'niebieski': 'zielony',
                 'zielony': 'pomaranczowy',
            'pomaranczowy': 'czerwony'}
        
        player = play.get(player_turn.get())
        return player           

    def consumption(self, player, updt='N'):
        
        print('consumption fn')
        consume = {0: 5,
                   1: 4, 2: 4, 3: 4, 4: 4,
                   5: 3, 6: 3, 7: 3, 8: 3,
                   9: 2, 10: 2, 11: 2, 12: 2,
                  13: 1, 14: 1, 15: 1, 16: 1}
        
        if updt == 'Y':
            new_consume = consume.get(player.yellow_dice.get() - (player.robotnicy.get()),0)
            player.zboze.set(player.zboze.get() - new_consume)    
        
        new_consume = consume.get(player.yellow_dice.get() - (player.robotnicy.get()),0)
        col = 'green' if new_consume < 1  else 'red'
        player.m_yellow.label.configure(text=new_consume, fg=col)

    def corruption(self, player, updt='N'):
        
        print('corruption fn')
        blue_dice_cnt = {0: 12,
                         1: 6, 2: 6, 3: 6, 4: 6, 5: 6,
                         6: 2, 7: 2, 8: 2, 9: 2, 10: 2}
        
        new_blue_dice_cnt = blue_dice_cnt.get(player.blue_dice.get() - (player.kamienie.get() + player.zboze.get()),0)
        col = 'green' if new_blue_dice_cnt < 1  else 'red'
        player.m_blue.label.configure(text=new_blue_dice_cnt, fg=col)
        if updt == 'Y':
            player.kamienie.set(player.kamienie.get() - new_blue_dice_cnt)      
        
    def buy_worker(self, player):
        
        print('buy_worker fn')
        work = {0: 7,
                1: 5, 2: 5, 3: 5, 4: 5,
                5: 4, 6: 4, 7: 4, 8: 4,
                9: 3, 10: 3, 11: 3, 12: 3}
        
        new_work = work.get(player.yellow_dice.get() - player.robotnicy.get(),2)
        player.zboze.set(player.zboze.get() - new_work)  
        player.robotnicy.set(player.robotnicy.get() + 1) 
            
        new_work = work.get(player.yellow_dice.get() - player.robotnicy.get(),2)
        player.m_robotnicy.button.configure(text = 'Buy (' + str(new_work) + ')') 
        
        self.cmd_wrpr('corruption')
        self.cmd_wrpr('consumption')
        
    def update_stats(self, player):     
        
        player.kultura.set(player.kultura.get() + player.g_kultura.get())
        player.nauka.set(player.nauka.get() + player.g_nauka.get())

    def update_wheat(self, player): 
        
        player.zboze.set(player.zboze.get() + player.g_zboze.get())     
        
    def update_stone(self, player): 
        
        player.kamienie.set(player.kamienie.get() + player.g_kamienie.get())
 
    def update_happynes(self, player):
        
        print('update_happynes fn')
        happynes = {1: 9, 
                    2: 13, 
                    3: 15, 
                    4: 17,
                    5: 19,
                    6: 21,
                    7: 23,
                    8: 25,
                    9: 27}
        
        new_happynes = happynes.get(player.zadowolenie.get(),7)
        print(new_happynes)
        
        col = 'green' if new_happynes > player.robotnicy.get()  else 'red'
        player.m_robotnicy.label.configure(text=new_happynes, fg=col)  
        
#         col = 'green' if new_happynes < player.zadowolenie.get()  else 'red'
#         player.m_zadowolenie.label.configure(text=new_happynes, fg=col)        
        
    def cmd_wrpr(self, cmd, attr='default', mod='none'):
        
        player = self.turn()
        
#         cmds = {'corruption': self.corruption,
#              blue_dice_color: self.corruption,
#                     'grey55': self.corruption,
#                       'gold': self.corruption,
#                'consumption': self.consumption,
#                     'orange': self.consumption,             
#                      'stats': self.update_stats,
#                      'stone': self.update_stone,
#                      'wheat': self.update_wheat,
#            zadowolenie_color: self.update_happynes,
#                     'worker': self.buy_worker}        
        
        cmds = {'corruption': self.corruption,
               'consumption': self.consumption,       
                     'stats': self.update_stats,
                    'worker': self.buy_worker}                
        
        new_cmd = cmds.get(cmd,'default')

        if mod != 'none':
            print('+ lub -')
            self.add(attr, mod)
        
        self.corruption(player)
        self.consumption(player)
        self.update_happynes(player)
        
        if new_cmd != 'default':
            print('mod')
            
        if cmd == 'worker':
            self.buy_worker(player)
        if cmd == 'corruption':
            self.corruption(player, 'Y')
            
        
        
#Attributes
class manipulator():        
    def __init__(self, set_x, set_y, attr, color, growth=0, growth_ind='N', label_ind='N', button_ind='N'):
        
        u = util()
        Entry(root, textvariable = attr, bd = 2, width=3, insertwidth = 1, font = 10, bg=color).place(x=set_x, y=set_y)
        Button(root, text= '\/', width=2, height=1, command = lambda: u.cmd_wrpr(color, attr, 'subtrack') ).place(x=set_x - 25, y=set_y)
        Button(root, text= '/\\', width=2, height=1, command = lambda: u.cmd_wrpr(color, attr, 'add') ).place(x=set_x + 41, y=set_y)
        
        if growth_ind == 'Y':
            Entry(root, textvariable = growth, bd = 2, width=2, insertwidth = 1, font = 10, bg=color).place(x=set_x + 100, y=set_y)
            Button(root, text= '-', width=2, height=1, command = lambda: u.cmd_wrpr(growth, 'subtrack') ).place(x=set_x +100 - 25, y=set_y)
            Button(root, text= '+', width=2, height=1, command = lambda: u.cmd_wrpr(growth, 'add') ).place(x=set_x + 100 + 30, y=set_y)        
        
        self.label = Label(root, text='0', font=("Helvetica", 14), fg='green', bg=background)
        if label_ind == 'Y':    
            self.label.place(x=set_x + 125, y=set_y)             
        
        self.button = Button(root, text= 'Buy (2)', width=5, height=1, bg="orange", state=button_state.get(), command = lambda: up.cmd_wrpr('worker'))
        if button_ind == 'Y':    
            self.button.place(x=set_x + 75, y=set_y)              
        
#Players
class player_class():
    def __init__(self, set_x, set_y, player_name, color):
        
        self.player = StringVar()        
        if player_name == 'Czerwony':
            self.player.set('Czerwony')
            
        self.kultura = IntVar()
        self.g_kultura = IntVar()        
        self.nauka = IntVar()
        self.g_nauka = IntVar()
        self.g_nauka.set(1)
        sila = IntVar()
        sila.set(1)
        
        cywilne = IntVar()
        cywilne.set(4)
        militarne = IntVar()
        militarne.set(3)
        
        self.blue_dice = IntVar()
        self.blue_dice.set(16)
        self.kamienie = IntVar()
        self.g_kamienie = IntVar()  
        self.g_kamienie.set(2)
        
        self.yellow_dice = IntVar()
        self.yellow_dice.set(25)
        self.zboze = IntVar()
        self.g_zboze = IntVar()
        self.g_zboze.set(2)
        
        self.robotnicy = IntVar()
        self.robotnicy.set(7)
        self.zadowolenie = IntVar()        
        
        manipulator(set_x, set_y, self.kultura, 'dark orchid', growth=self.g_kultura, growth_ind='Y')         
        manipulator(set_x, set_y + 40 , self.nauka, 'dodger blue', growth=self.g_nauka, growth_ind='Y') 
        manipulator(set_x, set_y + 80 , sila, 'indian red')         
        
        manipulator(set_x, set_y + 160, cywilne, 'white')
        manipulator(set_x, set_y + 200, militarne, 'orange red')
        
        m_blue = manipulator(set_x, set_y + 280, self.blue_dice, 'blue', label_ind='Y')       
        self.m_blue = m_blue
        manipulator(set_x, set_y + 320, self.kamienie, 'grey55', growth=self.g_kamienie, growth_ind='Y')         
        manipulator(set_x, set_y + 360, self.yellow_dice, 'orange', label_ind='Y')
        m_yellow = manipulator(set_x, set_y + 400 , self.zboze, 'gold', growth=self.g_zboze, growth_ind='Y')   
        self.m_yellow = m_yellow
        m_robotnicy = manipulator(set_x, set_y + 440, self.robotnicy, 'orange', label_ind='Y', button_ind='Y')
        self.m_robotnicy = m_robotnicy
        self.m_robotnicy.label.configure(text = '7')
        m_zadowolenie = manipulator(set_x, set_y + 480, self.zadowolenie, 'dark orange')   
        self.m_zadowolenie = m_zadowolenie
        
        if player_name == 'Czerwony':
            m_robotnicy.button.configure(state = 'normal')    

czerwony = player_class(210, 40, 'Czerwony', 'red')
niebieski = player_class(420, 40, 'Niebieski', 'blue')    
zielony = player_class(630, 40, 'Zielony', 'green')
pomaranczowy = player_class(840, 40, 'PomaraÅ„czowy', 'orange')

up = util()
root.mainloop()
```
